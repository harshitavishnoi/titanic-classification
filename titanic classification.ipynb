# One-Hot encodings

```python
#pd.get_dummies(train_data['Sex'])
```

```python
train_data['Sex'] = train_data['Sex'].map({'male': 0, 'female': 1})
train_data['Embarked'] = train_data['Embarked'].map({'S': 0, 'C': 0.5, 'Q': 1})
```

```python
train_data
```

# splitting the dataset into train and test

```python
Y_train = train_data["Survived"]

features = ["Fare","Pclass", "Sex", "SibSp", "Parch", "Embarked"]

X_train = pd.get_dummies(train_data[features]) #Convert categorical variable into dummy/indicator variables.

X_train
```

```python
from sklearn.model_selection import train_test_split

X_trainv, X_valid, Y_trainv, Y_valid = train_test_split(X_train, Y_train, test_size=0.2, random_state=5)

print("X_trainv_size =", X_trainv.shape, '\n', "X_valid_size =", X_valid.shape)
```

# building the model

```python
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.model_selection import GridSearchCV

model = GradientBoostingClassifier(max_depth=1)
n_estimators = [100,1000,5000,10000]
learning_rate = [1,0.1,0.01,0.001]
params = { 'n_estimators': n_estimators, 'learning_rate': learning_rate }

boost_grid = GridSearchCV(estimator = model, param_grid = params, 
                                cv = 5, verbose=2, scoring='accuracy',n_jobs = -1)

boost_grid.fit(X_trainv,Y_trainv)

print(boost_grid.best_params_)
```

# Training the model

```python
model=GradientBoostingClassifier(n_estimators=1000,max_depth=1,learning_rate=1)
model.fit(X_trainv,Y_trainv)
y_pred=model.predict(X_valid)

from sklearn.metrics import accuracy_score
print('Accuracy:', accuracy_score(Y_valid, y_pred))
```

```python
model.fit(X_train,Y_train)
```
